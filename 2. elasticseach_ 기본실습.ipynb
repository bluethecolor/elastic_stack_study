##### 실습 1: 인덱스 생성 및 데이터 삽입
인덱스 생성:
```json
PUT /test_index
{
  "settings": {
    "number_of_shards": 1,
    "number_of_replicas": 0
  }
}
``````

- PUT 명령으로 'test_index'라는 이름의 인덱스를 생성합니다.
- 샤드(shards)와 복제본(replicas)의 수를 설정할 수 있습니다.

데이터 삽입:
```json
POST /test_index/_doc/1
{
  "name": "John Doe",
  "age": 30,
  "interests": ["football", "music"]
}
``````

- POST 명령으로 'test_index' 인덱스에 문서를 삽입합니다.
- 문서에는 'name', 'age', 'interests' 필드가 포함됩니다.
  
##### 실습 2: 데이터 검색
전체 데이터 검색:

```json
GET /test_index/_search
{
  "query": {
    "match_all": {}
  }
}
``````
- GET 명령으로 'test_index' 인덱스의 모든 문서를 검색합니다.
  
  
특정 필드 검색:
```json
GET /test_index/_search
{
  "query": {
    "match": {
      "name": "John Doe"
    }
  }
}
``````
- 'name' 필드가 'John Doe'인 문서를 검색합니다.

##### 실습 3: 데이터 업데이트 및 삭제
데이터 업데이트:
```json
POST /test_index/_update/1
{
  "doc": {
    "age": 31
  }
}
``````
- ID가 1인 문서의 'age' 필드를 31로 업데이트합니다.
데이터 삭제:
```json
DELETE /test_index/_doc/1
``````
- ID가 1인 문서를 삭제합니다.
##### 4단계: 집계 및 분석
- Elasticsearch는 데이터 분석을 위한 강력한 집계 기능을 제공합니다.
- 예를 들어, 다음 집계 쿼리는 'age' 필드의 평균값을 계산합니다:
```json
GET /test_index/_search
{
  "size": 0,
  "aggs": {
    "average_age": {
      "avg": {
        "field": "age"
``````
